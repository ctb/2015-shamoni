```python
s17 = numpy.loadtxt('SML_17.dist', skiprows=1, delimiter=',')
s18 = numpy.loadtxt('SML_18.dist', skiprows=1, delimiter=',')
s19 = numpy.loadtxt('SML_19.dist', skiprows=1, delimiter=',')
```

```python
plot(s17[:,0], s17[:,1], label='17')
plot(s18[:,0], s18[:,1], label='18')
plot(s19[:,0], s19[:,1], label='19')
axis(xmax=200, ymax=1e4)
legend(loc='upper right')
```
